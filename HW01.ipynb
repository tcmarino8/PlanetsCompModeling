```python
import numpy as np
import matplotlib.pyplot as plt
```

#1 Below

```python
a = 1
b = a
a *= 2
print(a)
print(b)

```

Things to note about the above script is that when you set b to a and then modify a it does not modify b.

#2 below

```python
a = np.array( [ 1, 2, 3, 3, 5, 5 ] )
b = a.copy()
a *=2
print(a)
print(b)
```

So at first the b is modified when a is changed if a is set to b. As in that b and a point to the same list, however to make a change you can do .copy to create a new list a with a pointer.

#4 below

```python
x = np.linspace(0, 2*np.pi, 360)
fx = 3*np.cos(x)
gx = 2*np.cos(2*x)
hx = np.cos(3*x)
x = np.degrees(x)

fig = plt.figure(figsize = (20, 14))
plt.plot(x, fx, '--', label = '3cos(x)')
plt.plot(x, gx, 'D-', label = '2cos(2x)')
plt.plot(x, hx, 'P:', label = 'cos(3x)')
plt.xlabel('Degree')
plt.ylabel('Amplitude')
plt.title('Cosine variations')
plt.legend()

plt.show()
```

#5 Below

```python
r = [.5, 1.5, 2.5, 3.5, 4]
x = np.linspace(0, 1, 200)
plt.plot(x, x, label = 'f(x)')
def g(x, val):
    return x*val*(1-x)
for val in r:
    plt.plot(x, g(x, val), label=f'g(x), r={val}')
plt.legend()
plt.show()
```

First, the graph with r = .5 never intersects the linear graph, however all the others do. Once r = 2.5 and beyond, it intersects with the linear line after the maximum so as it is decreasing. They all take the same shape just different levels of arc.

#6 Below

```python
def h(x, val):
    return val*g(x, val)*1-g(x,val)

for val in r:
    plt.plot(x, h(x, val), label = f'H(x), r={val}')
    plt.plot(x, g(x, val), label=f'g(x), r={val}')
plt.legend()
plt.show()
```
